```python
import pandas as pd
```

contratos = pd.read_csv('tabela_contratos.csv')

```python
contratos
```

```python
datas = pd.read_csv('tabela_datas.csv')
```

```python
datas
```

```python
empresas = pd.read_csv('tabela_empresas.csv')
```

```python
empresas
```

```python
contratos_emp = contratos.merge(empresas, left_on='fk_empresa_contratada', right_on='id_empresa', how='left')
```

```python
contratos_emp
```

```python
contratos_emp.drop(columns=['id_empresa', 'fk_empresa_contratada'], inplace=True)
```

```python
contratos_emp.head(10)
```

```python
contratos_emp_data = contratos_emp.merge(datas, left_on='inicio_vigencia', right_on='id_data', how='left')
```

```python
contratos_emp_data.head(10)
```

```python
contratos_emp_data.drop(columns=['inicio_vigencia', 'id_data'], inplace=True)
```

```python
contratos_emp_data.rename(columns={'data': 'data_inicio_vingencia'}, inplace=True)
```

```python
contratos_emp_data.head(5)
```

```python
contratos_inicio_fim = contratos_emp_data.merge(datas, left_on='termino_vigencia', right_on='id_data', how='left')
```

```python
contratos_inicio_fim.drop(columns=['termino_vigencia', 'id_data'], inplace=True)
```

```python
contratos_inicio_fim.rename(columns={'data': 'data_termino_vingencia'}, inplace=True)
```

```python
contratos_inicio_fim.head(5)
```

```python
contratos_inicio_fim.count()
```

```python
contratos_inicio_fim.dtypes
```

```python
contratos_inicio_fim.data_inicio_vingencia = pd.to_datetime(contratos_inicio_fim.data_inicio_vingencia, format='%d/%m/%Y').dt.date
```

```python
# existe uma data inválida, o código abaixo serve para achar qual data está errada e alterar.
for i in contratos_inicio_fim.data_termino_vingencia:
    print(i)
    print(pd.to_datetime(i))
```

```python
contratos_inicio_fim.data_termino_vingencia = contratos_inicio_fim.data_termino_vingencia.str.replace('31/09/2017', '30/09/2017')
```

```python
contratos_inicio_fim.data_termino_vingencia = pd.to_datetime(contratos_inicio_fim.data_termino_vingencia, format='%d/%m/%Y').dt.date
```

```python
contratos_inicio_fim.dtypes
```

```python
contratos_inicio_fim.head(5)
```

```python
contratos_inicio_fim['dias_contrato'] = (contratos_inicio_fim.data_termino_vingencia - contratos_inicio_fim.data_inicio_vingencia)
```
